#### Cleaning text data with NLTK 1

```python
import nltk
import re
```

```python
##download stop words
nltk.download("stopwords")
```

```python
##import stop words module
from nltk.corpus import stopwords
```

```python
res_data['Review'][0]
```

```python
### to keep only alphabetic charecter , regular expression library
fresh_review_data= re.sub('[^a-zA-Z]',' ',res_data['Review'][0])
```

```python
fresh_review_data
```

```python
fresh_review_data= fresh_review_data.split()
```

```python
fresh_review_data
```

```python
stopwords.words('english')
```

```python
preview=[]
for word in fresh_review_data:
  if word not in stopwords.words('english'):
    preview.append(word)
    
```

```python
preview
```

```python
review = [word for word in fresh_review_data if word not in stopwords.words('english')]
```

#### Cleaning text data with NLTK 2

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
review = [ps.stem(word) for word in review]
```

```python
review
```

```python
review = " ".join(review)
```
